```python

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
dataframe=pd.read_csv("/content/Titanic-Dataset.csv")
dataframe.head(10)
```

```python
dataframe.describe()
```

```python
dataset = dataframe['Fare'].dropna()
plt.figure(figsize=(15, 10))
plt.hist(dataset, bins=30, color='blue', edgecolor='black')
plt.title('Histogram for Fare')
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
X = dataframe.select_dtypes(include='number').dropna()
fig, ax = plt.subplots(figsize=(15, 10))
sns.boxplot(data=X, ax=ax)
ax.set_title("Boxplot of Numerical Features Before Removing Outliers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
sns.pairplot(dataframe[['Survived', 'Pclass', 'Age']])
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(dataframe.select_dtypes(include=['int64', 'float64']).corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```

```python
dataframe['FamilySize'] = dataframe['SibSp'] + dataframe['Parch']
sns.countplot(data=dataframe, x='FamilySize', hue='Survived')
plt.title("Survival by Family Size")
plt.show()
```

```python
dataframe['FamilySize'] = dataframe['SibSp'] + dataframe['Parch']
plt.figure(figsize=(7, 4))
sns.countplot(data=dataframe, x='FamilySize', hue='Survived')
plt.title("Survival by Family Size")
plt.show()
```
